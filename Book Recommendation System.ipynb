# Popularity Based Recommendation System 

```python
# Merge ratings and books dataframes
```

```python
df = ratings.merge(books, on='ISBN')
```

```python
num_df = df.groupby('Book-Title')['Book-Rating'].count().reset_index()
```

```python
num_df.rename(columns = {'Book-Rating':'Num-Rating'},inplace=True)
```

```python
num_df
```

```python
avg_df = df.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_df.rename(columns = {'Book-Rating':'avg-Rating'},inplace=True)
avg_df
```

```python
popularity_df = avg_df.merge(num_df, on='Book-Title')
```

```python
popularity_df = popularity_df[popularity_df['Num-Rating']>450].sort_values('avg-Rating', ascending=False)
```

```python
popularity_df = popularity_df.head(50)
```

```python
popularity_index = popularity_df.merge(books, on='Book-Title')
```

```python
popularity_index = popularity_index.drop_duplicates('Book-Title')
```

```python
popularity_index.shape
```

```python
popularity_rs = popularity_index[['Book-Title','Book-Author','Num-Rating', 'avg-Rating', 'Image-URL-M']]
```

# Collaborative Filtering Based Reccomendation System

```python
Ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
x = Ratings_with_name.groupby('User-ID')['Book-Rating'].count() > 200
```

```python
user_index = x[x].index
```

```python
filtered_rating = Ratings_with_name[Ratings_with_name['User-ID'].isin(user_index)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] > 50
```

```python
movie_index = y[y].index
```

```python
filtered_movie = filtered_rating[Ratings_with_name['Book-Title'].isin(movie_index)]
```

```python
filtered_movie
```

```python
pt = filtered_movie.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace =True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(pt)
```

```python
similarity.shape
```

```python
def recommendation(book):
    book_index = np.where(pt.index == book)[0][0]
    similar_items = sorted(list(enumerate(similarity[book_index])), key= lambda x:x[1], reverse=True)[1:6]
    suggested_index = list(map(lambda x: x[0], similar_items))
    my_list = [x for x in pt.index[suggested_index]]
    print(my_list)
```

```python
recommendation('A Bend in the Road')
```
